```python
import matplotlib.pyplot as plt

def plotCounts(fn, title, scale, shrink):
    counts = []
    with open(fn) as f:
        for line in f:
            tokens = line.split(', ')
            for token in tokens:
                counts.append(int(token))
    print(len(counts))
    plt.hist(counts, bins = [i for i in range(0, max(counts)+scale-1, scale)])
    plt.xlim(0, max(counts)/shrink)
    plt.title(title)
    plt.xlabel('Number of contained reads')
    plt.show()
```

```python
plotCounts('humancountcontaining.txt', 'Human', 10, 2)
```

```python
plotCounts('arabacountcontaining.txt', 'Arabidopsis', 2, 20)
```
